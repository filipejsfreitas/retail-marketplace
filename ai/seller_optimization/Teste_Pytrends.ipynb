# Popularidade de um produto

```python
from pytrends.request import TrendReq
from pytrends.exceptions import ResponseError
from pytrends import dailydata
```

```python
pytrends = TrendReq(hl='en-UK')
```

```python
pytrends.build_payload(kw_list=['fridge'
,'coffee machine'])
```

```python
df = pytrends.interest_over_time()
df
```

```python
interessante = df.iloc[-2:]
interessante
```

```python
diferenca = abs(interessante['coffee machine'][0] - interessante['coffee machine'][1])
print(diferenca)
if interessante['fridge'][0] >interessante['fridge'][1]:
    print("Produto tem mais popularidade. Sendo assim, é possivel ser necessario aumentar stock.")
    inflacao = diferenca/100 * 20
    print("Novo preço : ", inflacao + 20)
else:
    print("Produto diminui de popularidade.")
```

```python
def gtrends_region(kw_list, key_ref, save_name="", 
                   category=0, time='all', loc='', res='COUNTRY'):
    #iterate every 4 item in list plus a keyword reference as the relative comparison
    i = 0
    while i < len(kw_list):
        l = kw_list[i:(i+4)]
        l.append(key_ref)
        pytrends.build_payload(l, cat=category, timeframe=time, geo=loc, gprop='')
        # resolution can be set to 'REGION' when loc is set to specific country, e.g. 'US'
        df_region = pytrends.interest_by_region(resolution=res, inc_low_vol=True, inc_geo_code=False)
        df_region.reset_index(inplace=True)
        df_region_name = "gtrends_region"+str(save_name)+str((i+4)//4)+".csv"
        df_region.to_csv(df_region_name, index = False)
        i += 4
        
gtrends_region(kw_list, 'Noodles', "_worldwide_",
               category=71, time='all', loc='', res='COUNTRY')
```
